```python
import matplotlib.pyplot as plt

# Datos para el gráfico de dispersión
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Crear el gráfico de dispersión
plt.scatter(x, y)

# Etiquetas y título
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Gráfico de dispersión simple")

# Mostrar el gráfico
plt.show()
```
